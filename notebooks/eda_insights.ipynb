```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('processed_global_data.csv')

# Distribution of Spending
plt.figure(figsize=(10, 5))
sns.histplot(df['Total_Spend'], bins=50, kde=True, color='purple')
plt.title('Global Spending Distribution')
plt.show()

# Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df[['Purchase_Frequency', 'Average_Order_Value', 'Customer_LTV']].corr(), annot=True, cmap='RdYlGn')
plt.show()
```
