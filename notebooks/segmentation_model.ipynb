```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

df = pd.read_csv('processed_global_data.csv')
features = ['Purchase_Frequency', 'Average_Order_Value', 'Last_Purchase_Days_Ago']
x = StandardScaler().fit_transform(df[features])

# PCA for Dimensionality Reduction
pca = PCA(n_components=2)
pc_data = pca.fit_transform(x)

# Clustering
kmeans = KMeans(n_clusters=5, random_state=42).fit(pc_data)
df['Cluster'] = kmeans.labels_

print("Clustering complete. Personas identified.")
```
